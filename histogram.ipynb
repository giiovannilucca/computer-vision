# Histogramas: teoria e prática.

<p align='justify'> O histograma na área de Processamento de Imagens representa a distribuição dos valores dos pixels de uma imagem digital, podendo ser aplicado em diversas tarefas, tais como o aumento de contraste, segmentação do objeto, extração de características e recuperação de imagens. 

Outro ponto importante, o histograma fornece uma indicação da qualidade da imagem quanto ao contraste e intensidade luminosa. Como características importantes do histograma, podemos citar:
* As informações espaciais dos objetos contidos na imagem não são representadas no histograma
* Um histograma é único para um determinada imagem, mas o inverso não é verdadeiro
* A movimentação de objetos (translação e rotação) em uma imagem não tem efeito sobre o seu histograma

Em relação ao aumento do contraste, existe uma técnica chamada de equalização do histograma. Esta técnica tem como objetivo espalhar as ocorrências dos pixels sobre todos os níveis de cinza, melhorando assim, o contraste da imagem de entrada.<p>

```python
#importando as bibliotecas necessárias
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
#lendo uma imagem de tomografia ocular (OCT)
input_image = cv.imread('oct.jpg', 0)
input_image = cv.resize(input_image, dsize = None, fx = 0.5, fy = 0.5) #reduzindo a imagem pela metade do tamanho original
cv2_imshow(input_image)
```

```python
#plotando o histograma original da imagem de entrada
hist = cv.calcHist(input_image, [0], None, [256], [0, 256])

plt.plot(hist)
```

```python
#aplicando a técnica de equalização do histograma na imagem de entrada
equalized_image = cv.equalizeHist(input_image)

images = np.hstack([input_image, equalized_image])
cv2_imshow(images)
```

```python
#plotando o histograma equalizado da imagem de entrada
hist = cv.calcHist(equalized_image, [0], None, [256], [0, 256])

plt.plot(hist)
```
