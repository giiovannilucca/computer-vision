# Filtragem de imagens: teoria e prática.

<p align='justify'>As operações de filtragem podem ser realizadas tanto no domínio do espaço quanto de frequência. Além disso, os filtros podem ser classificados em três categorias: passa-baixas, passa-altas e passa-faixa.<p>

<p align='justify'>Em resumo, um filtro passa-baixa tem como objetivo suavizar a imagem de entrada, uma vez que as altas frequências, que correspondem às transições abruptas, são atenuadas. Por outro lado, um filtro passa-alta tem como objetivo tornar mais nítida as transições entre regiões diferentes, conhecidas como bordas.<p>

<p align='justify'>No domínio espacial, o nível de cinza de um ponto f(x,y) após a transformação depende do valor do nível de cinza original do ponto e de outros pontos da vizinhança de f(x,y) usando matrizes denominadas de máscaras ou kernels. A cada posição da máscara está associado um valor numérico, chamado de coeficiente.<p>

Os filtros passa-baixas mais conhecidos e utilizados são:

* Filtro da média: Cada pixel é substituído pelo valor médio dos pixels da vizinhança.

* Filtro Gaussiano: Cada pixel é substituído por uma média ponderada dos pixels da vizinhança.

* Filtro da mediana: Cada pixel é substituído pela mediana dos pixels da vizinhança. 

* Filtro bilateral: Cada pixel é substituído por uma média ponderada dos pixels da vizinhança com preservação de bordas.

```python
#importando as bibliotecas necessárias
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#lendo uma imagem de mamografia digital com a presença de um nódulo
input_image = cv.imread('mamografia.png', 0)
input_image = cv.resize(input_image, dsize = None, fx = 0.5, fy = 0.5) #reduzindo a imagem pela metade do tamanho original
cv2_imshow(input_image)
```

```python
#Filtro da média
mean_image = cv.blur(input_image, (9,9)) #máscara de tamanho 9x9

gaussian_image = cv.GaussianBlur(input_image, (9,9), 0) #máscara de tamanho 9x9

median_image = cv.medianBlur(input_image, 9) #máscara de tamanho 9x9

bilateral_image = cv.bilateralFilter(input_image, 9, 21, 21) #máscara de tamanho 9x9

images = np.hstack((input_image, mean_image, gaussian_image, median_image, bilateral_image))

cv2_imshow(images)
```
